```python
import math


sample_sizes = np.linspace(10, 100000, 5000, dtype=np.int32)

moments = {'mean': [], 'std': []}
bounds = {'mean': [], 'std': []}
for sample_size in sample_sizes:
    # sampling
    sample = np.random.normal(mean, std, sample_size)

    moments['mean'].append(sample.mean())
    moments['std'].append(sample.std(ddof=1))

    z_critical = stats.t.ppf(df=sample_size - 1, q=1 - alpha/2) 
    mean_margin = z_critical * (sample.std(ddof=1)/math.sqrt(sample_size))
    mean_bounds = [sample.mean() - mean_margin, sample.mean() + mean_margin]
    mean_bounds = np.array(mean_bounds)
    bounds['mean'].append(mean_bounds)

    z_critical_right = stats.chi2.ppf(1 - alpha/2, df=sample_size - 1)
    z_critical_left = stats.chi2.ppf(alpha/2, df=sample_size - 1)
    right_bound = np.sqrt(sample_size - 1)*sample.std(ddof=1) / np.sqrt(z_critical_right)
    left_bound = np.sqrt(sample_size - 1)*sample.std(ddof=1) / np.sqrt(z_critical_left)
    std_bounds = [left_bound, right_bound]
    std_bounds = np.array(std_bounds)
    bounds['std'].append(std_bounds)
    
moments['mean'] = np.array(moments['mean'])
moments['std'] = np.array(moments['std'])
bounds['mean'] = np.array(bounds['mean'])
bounds['std'] = np.array(bounds['std'])
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.set_title('Empirical evidence of the law of large numbers')
ax.fill_between(sample_sizes, bounds['mean'][:, 0], bounds['mean'][:, 1], alpha=0.2, color='green')
_ = sns.lineplot(sample_sizes, moments['mean'], label='Mean')
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.set_title('Empirical evidence of the law of large numbers')
ax.fill_between(sample_sizes, bounds['mean'][:, 0], bounds['mean'][:, 1], alpha=0.2, color='green')
_ = sns.lineplot(sample_sizes, moments['std'], label='Std')
```

From law of large numbers we can conduct that our mean and variance is converting to the almost the same number.

```python
# simulating data
sample = np.random.standard_t(df=5, size=100)

# defining likelihood
def likelihood(data, df):
    return -np.sum(np.array(list(map(lambda x: np.log(stats.t.pdf(x, df)), data))))

# optimizing likelihood with help of scipy optimizer
res = scipy.optimize.minimize(lambda x: likelihood(sample, x), x0=12., constraints={'type': 'ineq', 'fun': lambda x: x})
print('Optimal value: {}'.format(res['x'][0]))
```

```python
values = np.linspace(3, 15, 15)
res = [likelihood(sample, df=val) for val in values]
```

```python
def CDF(mean=0, std_dev=1, x_left=-4, x_right=4, width=0.0001):
    CDF = 0
 
    ''' Call PDF for each value of x_left to x_right '''
    # calculate number of panels (+ 1 includes x_right)
    the_range = int((x_right - x_left) / width) + 1
    for i in range(the_range):
        x = x_left + i * width  # current x value
        y = PDF(x, mean, std_dev)  # probability of this panel
        panel = y * width  # this panel area under PDF curve
        CDF += panel  # sum panel areas = CDF
       
    return CDF
```

```python
import math
def PDF(x, mean, std_dev):
    probability = 1.0 / math.sqrt(2 * 3.141592*(std_dev)**2)  # first part of equation
    probability *= math.exp(-0.5 * ((x - mean)/std_dev)**2)  # multiply first part to second part
 
 
    return probability
```

**T-test formula**

x̄ = Observed Mean of the Sample

μ = Theoretical Mean of the Population

s = Standard Deviation of the Sample

n = Sample Size

**Data**

μ0 = 180000

Test, whether mean, is equal to μ0

**Hypothesis**

Null hypothesis H0: μ = μ0

Alternative hypothesis H1: μ > μ0

```python
from sympy import *
from IPython.display import display, Math

init_printing()
x, z, μ, s, n, x̄ = symbols("x z μ s n x̄")

x = (x̄ - μ) / (s / sqrt(n))
print('Formula')
display(Math('x = '+latex(x)))

print('Formula with inserted values')
print()
x, z, μ, s, n, x̄ = symbols(f"x z 180000 {data.SalePrice.std()} {len(data.SalePrice)} {data.SalePrice.mean()}")
x = (x̄ - μ) / (s / sqrt(n))
display(Math('x = '+latex(x)))
```

```python
from scipy import stats


t_stat, p_value = stats.ttest_1samp(data.SalePrice, popmean=180000)
print(f"t Statistic: {t_stat}")
print(f"P Value: {p_value}")
print(f"One tailed P value: {p_value}")
df = len(data.SalePrice) - 1
alpha = 0.05
critical_value = stats.t.ppf(1.0 - alpha, df)
print(f'Critical value: {critical_value}')
if abs(t_stat) >= critical_value:
    print('Accept H1 hypothesis')
else:
    print('Accept H0 hypothesis')

if p_value <= alpha:
    print('Accept H1 hypothesis')
else:
    print('Accept H0 hypothesis')
```

We accept H0 hypothesis, because 0.443 < 1.645. We have statistically significant evidence at alpha = 0.05.
